# Exploración Inicial: Premier League

Este notebook está diseñado para ser flexible. Cada sección intentará localizar los datos automáticamente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json

# Configuración de visualización
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Función para encontrar la carpeta de datos
def get_data_path():
    cwd = os.getcwd()
    base = os.path.dirname(cwd) if 'notebooks' in cwd else cwd
    return os.path.join(base, 'data', 'raw')

data_raw_dir = get_data_path()
print(f"Carpeta de datos: {data_raw_dir}")
```

## 1. Carga de Datos CSV
Cargamos los archivos históricos de la Premier League.

```python
import os, pandas as pd
path_dir = os.path.join(os.path.dirname(os.getcwd()) if 'notebooks' in os.getcwd() else os.getcwd(), 'data', 'raw')

files = {
    '24/25': os.path.join(path_dir, 'pl_2425.csv'),
    '23/24': os.path.join(path_dir, 'pl_2324.csv'),
    '22/23': os.path.join(path_dir, 'pl_2223.csv')
}

dfs = {}
for season, p in files.items():
    if os.path.exists(p):
        dfs[season] = pd.read_csv(p)
        print(f"✅ {season}: Cargada.")
    else:
        print(f"❌ {season}: No encontrada.")
```

## 2. Análisis de Resultados
¿Quién gana más, el local o el visitante?

```python
if '24/25' in dfs:
    df = dfs['24/25']
    df['Result_Label'] = df['FTR'].map({'H': 'Home Win', 'D': 'Draw', 'A': 'Away Win'})
    sns.countplot(data=df, x='Result_Label', palette='viridis')
    plt.title('Distribución de Resultados - Temporada 24/25')
    plt.show()
```

## 3. Exploración de Datos JSON (API)
Convertimos el archivo JSON de la API en una tabla plana y lo guardamos como CSV.

```python
import os, json, pandas as pd

current_dir = os.getcwd()
base = os.path.dirname(current_dir) if 'notebooks' in current_dir else current_dir
path_json = os.path.join(base, 'data', 'raw', 'api_sample_matches.json')
path_csv_out = os.path.join(base, 'data', 'raw', 'api_sample_matches.csv')

if os.path.exists(path_json):
    with open(path_json, 'r', encoding='utf-8') as f:
        api_data = json.load(f)

    df_api = pd.json_normalize(api_data['matches'])
    
    # GUARDAR COMO CSV
    df_api.to_csv(path_csv_out, index=False, encoding='utf-8')
    print(f"✅ JSON convertido y guardado como CSV en: {path_csv_out}")
    
    # Mostrar muestra
    display(df_api[['utcDate', 'homeTeam.name', 'awayTeam.name', 'score.fullTime.home', 'score.fullTime.away']].head())
else:
    print(f"❌ Archivo JSON no encontrado en: {path_json}")
```
